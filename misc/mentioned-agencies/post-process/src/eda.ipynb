```python
import ast

import pandas as pd
```

```python
def read_tbl():
    dfa = pd.read_csv("../data/output/mentioned_agencies_output.csv")
    dfb = pd.read_csv("../data/output/mentioned_agencies_corrections.csv")
    return dfa, dfb


def standardize_agencies(value):
    if pd.isna(value):
        return value

    # If it's a string representation of a list
    if isinstance(value, str) and value.startswith("["):
        try:
            # Safely evaluate the list string
            agencies_list = ast.literal_eval(value)
            return ", ".join(agencies_list)
        except Exception:
            return value

    # If it's already a plain string, return as-is
    return value


dfa, dfb = read_tbl()

dfb_correction_cases = [x for x in dfb["provisional_case_name"]]

dfa = dfa[~(dfa.provisional_case_name.isin(dfb_correction_cases))]

print(dfa)

df = pd.concat([dfa, dfb])

df["mentioned_agencies"] = df["mentioned_agencies"].apply(standardize_agencies)

# df = df[["provisional_case_name", "mentioned_agencies"]]


# df = df[~((df.mentioned_agencies.fillna("") == ""))]

df
# df.to_csv("../data/output/mentioned_agencies_2_5_2026.csv", index=False)
```
