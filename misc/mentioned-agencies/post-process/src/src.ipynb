```python
import pandas as pd
from pathlib import Path
```

```python
def combine_autofolio_csvs(input_dir="../data/input", output_path=None):
    """
    Read all CSVs from input_dir, extract specified columns, and combine into one DataFrame.
    """
    columns = ['provisional_case_name', 'mentioned_agencies', 'agency_name']
    input_path = Path(input_dir)
    output_path = Path("../data/output/mentioned_agencies_output.csv")


    if not input_path.exists():
        raise FileNotFoundError(f"Directory not found: {input_dir}")
    
    csv_files = list(input_path.glob("*.csv"))
    
    # Check if any CSV files found
    if not csv_files:
        raise FileNotFoundError(f"No CSV files found in {input_dir}")
    
    print(f"Found {len(csv_files)} CSV files")
    
    dfs = []
    for csv_file in csv_files:
        print(f"Reading: {csv_file.name}")
        df = pd.read_csv(csv_file, usecols=columns)
        dfs.append(df)
    
    combined_df = pd.concat(dfs, ignore_index=True)
    print(f"Combined shape: {combined_df.shape}")
    
    if output_path:
        combined_df.to_csv(output_path, index=False)
        print(f"Saved to: {output_path}")
    
    return combined_df

df = combine_autofolio_csvs()
```
